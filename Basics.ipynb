# Gender Gap in US College Majors (1970 to 2012)

The US Department of Education Statistics in releases a data set annually containing the percentage of bachelor's degrees granted to women. In this project, we will use data from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category. Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. You can download the dataset Randal compiled [here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv).

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

## Line Charts Comparing Gaps in All Majors
Here we want to visualise and compare the three categories of majors with a 6x3 plot. The main aim of this project is first to practice making plots and also reducing clutter and distractions in our visualisations by reducing the data-ink ratio.

While an easier way is to combine all three lists in the right order and write a single loop with range(0,18), I treated this as a mental exercise using ranges and simple division.

```python
fig = plt.figure(figsize=(18, 18))

#stem has 7 majors, for column 1
for sp in range(1,18,3):
    ax = fig.add_subplot(6,3,sp)
    maj_int = int(sp/3)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[maj_int]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[maj_int]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[maj_int])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    ax.set_yticks([0,50,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
    
    if sp == 16:
        ax.tick_params(labelbottom='on')
    
    if sp == 1:
        ax.text(2006, 83, 'Women')
        ax.text(2006, 15, 'Men')

#liberal arts has 5 majors, for column 2
for sp in range(2,17,3):
    ax = fig.add_subplot(6,3,sp)
    maj_int = int(sp/3)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[maj_int]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[maj_int]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[maj_int])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    ax.set_yticks([0,50,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
    
    if sp == 14:
        ax.tick_params(labelbottom='on')
    
    if sp == 2:
        ax.text(2006, 80, 'Women')
        ax.text(2006, 18, 'Men')


#the other 6 majors, for column 3
for sp in range(3,21,3):
    ax = fig.add_subplot(6,3,sp)
    maj_int = int((sp-1)/3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[maj_int]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[maj_int]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[maj_int])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    ax.set_yticks([0,50,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.5)
    
    if sp == 18:
        ax.tick_params(labelbottom='on')
    
    if sp == 3:
        ax.text(2006, 90, 'Woman')
        ax.text(2006, 6, 'Men')
    
plt.savefig('gender_degrees.png')
plt.show()
```

## Key Findings

#### Stem Majors
> Interestingly, the gender gap in Psychology flipped at the start of the 70s, with more women getting into the field than men. The gap has continued to widen since then, and has held constant since roughly the 2000s through to 2010. With the talk about women in programming jobs lately, it is interesting to see than the gap between men and women widened even further in the early 2000s through to 2010. However, a closely related discipline, Math and Statistics have a healthly share of the sexes.

#### Lib. Art Majors
> Most liberal art majors are dominated by women, especially language related majors. It is interesting to note that in the 1980s, women started become the majority in the Communications and journalism major. 

#### Other Majors
> Education, health professions, and public administration majors are dominated by women, with health profession having the widest gap, comparable to Engineering in STEM. Business and agriculture majors on the other hand have healthy 50/50 ratio of men to women.
